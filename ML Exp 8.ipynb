```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('/content/Mall_Customers.csv')
```

```python
df.head()
```

```python
# Select relevant features (Annual Income and Spending Score)
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
```

```python
# Add the cluster labels to the original dataframe
df['DBSCAN_Cluster'] = dbscan_labels
```

```python
# Visualize the clusters
plt.figure(figsize=(10, 6))
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=dbscan_labels, cmap='viridis', s=50)
plt.title('DBSCAN Clustering: Income vs Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
# Analyze the number of clusters and noise points
num_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)  # exclude noise points
num_noise = list(dbscan_labels).count(-1)
```

```python
# Output the results
print(f"Number of clusters: {num_clusters}")
print(f"Number of noise points: {num_noise}")
```
