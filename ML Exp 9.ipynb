```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Step 1: Load and Explore the Digits Dataset
digits = load_digits()
X_digits = digits.data  # Feature matrix (64 features per image)
y_digits = digits.target  # Target labels (digits 0-9)
```

```python
# Step 2: Load and Explore the CSV (Mall_Customers.csv)
mall_customers = pd.read_csv("/content/Mall_Customers.csv")
```

```python
# Let's view the first few rows of the CSV to understand its structure
print("Mall Customers CSV Preview:")
print(mall_customers.head())
```

```python
# Assuming we want to perform PCA on 'Annual Income (k$)' and 'Spending Score (1-100)' columns:
# Select relevant features from the mall customers dataset
X_mall = mall_customers[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Standardize the data
scaler = StandardScaler()
X_mall_scaled = scaler.fit_transform(X_mall)
```

```python
# Apply PCA for dimensionality reduction (2 components)
pca_mall = PCA(n_components=2)
X_mall_pca = pca_mall.fit_transform(X_mall_scaled)
```

```python
# Step 3: Visualize PCA for the mall customers data
plt.figure(figsize=(10, 6))
plt.scatter(X_mall_pca[:, 0], X_mall_pca[:, 1], alpha=0.8)
plt.title('PCA of Mall Customers Dataset (2D)')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```

```python
# Step 4: Apply PCA on the Digits dataset (already scaled earlier)
scaler_digits = StandardScaler()
X_digits_scaled = scaler_digits.fit_transform(X_digits)
```

```python
pca_digits = PCA(n_components=2)  # 2D reduction for visualization
X_digits_pca = pca_digits.fit_transform(X_digits_scaled)
```

```python
# Step 5: Visualize PCA for Digits dataset
plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_digits_pca[:, 0], X_digits_pca[:, 1], c=y_digits, cmap='tab10', alpha=0.8)
plt.colorbar(scatter)
plt.title('PCA of Digits Dataset (2D)')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```

```python
# Step 6: Analyze the variance explained by the first two principal components
explained_variance_mall = pca_mall.explained_variance_ratio_
explained_variance_digits = pca_digits.explained_variance_ratio_
```

```python
print(f"Explained variance by the first two components (Mall Customers): {explained_variance_mall}")
print(f"Explained variance by the first two components (Digits Dataset): {explained_variance_digits}")
```
